```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings("ignore")
```

```python
pages=200
page_size=10
reviews=[]
date = []
recommended=[]

for i in range(1,pages+1):
  print(f"scraping page{i}")
  page=requests.get(f"https://www.airlinequality.com/airline-reviews/british-airways/page{i}")
  soup=BeautifulSoup(page.content,"html.parser")

  #Reviews

  for item in soup.find_all("div",class_="text_content"):
     reviews.append(item.text)
  # Date
  for item in soup.find_all("time"):
                date.append(item.text)
  # Recommended
  for item in soup.find_all('td', class_='review-rating-header recommended'):
                                             recommended_text = item.find_next_sibling('td').text.strip()
                                             recommended.append(recommended_text)
```

```python
df=pd.DataFrame({"reviews":reviews,"date":date,"recommende":recommended})
pd.set_option("display.max_rows",None)
print(df.head())
```

```python
df.to_csv("AVIS_CLIENT.csv", index=False, header=True,sep=';')
```
